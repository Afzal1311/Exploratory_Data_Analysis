# Univariate analysis for categories having ON_time_payment_df target=0

```python
#putting y axis as log and plotting the graph
def Uni_variate(df,col,title,hue =None):
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 9 + 5*len(temp.unique())
    fig.set_size_inches(width , 9)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
    plt.show()
```

```python
Uni_variate(ON_time_payment_df,col='Total_income_range',title='Gender wise income range distribution',hue='Gender')
```

# Points to be concluded from the above graph.

Female counts are higher than male.
Income range from 100000 to 200000 is having more number of credits.
This graph show that females are more than male in having credits for that range.
Very less count for income range 400000 and above.

```python
Uni_variate(ON_time_payment_df,col='NAME_INCOME_TYPE',title='Gender wise Income type',hue='Gender')
```

For income type working, commercial associate, and State Servant the number of credits are higher than others.
